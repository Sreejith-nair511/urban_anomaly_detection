# Smart Urban Anomaly Detection – EDA
Exploratory Data Analysis for Bangalore urban sensor data.

```python
import sys, os
sys.path.insert(0, os.path.join('..', 'src'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from generator import load_or_generate
sns.set_theme(style='whitegrid')

df = load_or_generate('../data/raw/urban_data.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df.head()
```

```python
print('Shape:', df.shape)
print('\nMissing values:\n', df.isnull().sum())
df.describe().round(2)
```

```python
# Anomaly rate
if 'is_anomaly' in df.columns:
    rate = df['is_anomaly'].mean() * 100
    print(f'Anomaly rate: {rate:.1f}%')
    df['is_anomaly'].value_counts().plot(kind='bar', color=['#4CAF50','#F44336'])
    plt.title('Anomaly vs Normal')
    plt.xticks([0,1], ['Normal','Anomaly'], rotation=0)
    plt.show()
```

```python
# Feature distributions
features = ['AQI','temperature','humidity','traffic_density','noise_level']
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
for ax, col in zip(axes.flat, features):
    sns.histplot(df[col], kde=True, ax=ax, color='steelblue')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[features].corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# AQI by zone
plt.figure(figsize=(12, 5))
zone_aqi = df.groupby('location_id')['AQI'].mean().sort_values(ascending=False)
zone_aqi.plot(kind='bar', color='tomato')
plt.title('Average AQI by Bangalore Zone')
plt.ylabel('Mean AQI')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Hourly traffic pattern
plt.figure(figsize=(10, 4))
df.groupby('hour')['traffic_density'].mean().plot(marker='o', color='darkorange')
plt.title('Average Traffic Density by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Traffic Density')
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```
